```python
import sympy as sym
import numpy as np
from scipy.optimize import fsolve


x = sym.Symbol('x')


def laguerre_poly(n):

    exp_x = sym.exp(x)
    exp_minus_x_x_n = sym.exp(-x) * x**n
    diff_n = sym.diff(exp_minus_x_x_n, x, n)
    L_n = exp_x / sym.factorial(n) * diff_n
    return L_n


x_range = (0, 100)


for n in range(20):
    L_n = laguerre_poly(n)
    L_n_func = sym.lambdify(x, L_n, 'numpy')
    def func(x_val):
        return L_n_func(x_val)
    initial_guesses = np.linspace(0, 100, 100)
    roots = []
    for guess in initial_guesses:
        root = fsolve(func, guess)[0]
        if x_range[0] <= root <= x_range[1] and not any(np.isclose(root, r, atol=1e-5) for r in roots):
            roots.append(root)

    roots.sort()


    print(f"Polinomio de Laguerre L_{n}(x): {L_n}")
    print(f"Raíces reales en el intervalo [0, 100] de L_{n}(x): {roots}")
    print()
```
