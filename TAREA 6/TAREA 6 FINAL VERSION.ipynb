```python
#1
def propiedades_probabilidad():

    P = {
        "empty": 0,  # P(∅)
        "A": 0.3,    # P(A)
        "B": 0.5,    # P(B)
        "A∩B": 0.2   # P(A ∩ B)
    }

    assert P["empty"] == 0, "Error en P(∅): no es igual a 0"

    P_complemento_A = 1 - P["A"]
    assert P_complemento_A == 0.7, "Error en P(A^c): cálculo incorrecto"


    if P["A∩B"] == P["A"]:
        P_B_menos_A = P["B"] - P["A∩B"]
        assert P["B"] == P["A"] + P_B_menos_A, "Error en P(B) = P(A) + P(B - A)"
    else:
        print("Advertencia: A no está contenido en B, omitiendo esta propiedad.")

    assert P["A"] <= 1, "Error en P(A): no cumple P(A) ≤ 1"

    if P["A∩B"] == P["A"]:
        assert P["A"] <= P["B"], "Error en P(A) ≤ P(B)"
    else:
        print("Advertencia: A no está contenido en B, omitiendo esta propiedad.")

    P["A∪B"] = P["A"] + P["B"] - P["A∩B"]
    assert P["A∪B"] == P["A"] + P["B"] - P["A∩B"], "Error en P(A ∪ B): cálculo incorrecto"

    P["C"] = 0.2
    P["A∩C"] = 0.1
    P["B∩C"] = 0.15
    P["A∩B∩C"] = 0.05
    P_union_ABC = (
        P["A"] + P["B"] + P["C"]
        - P["A∩B"]
        - P["A∩C"]
        - P["B∩C"]
        + P["A∩B∩C"]
    )
    print("P(A ∪ B ∪ C) =", P_union_ABC)

    P_A_menos_B = P["A"] - P["A∩B"]
    print("P(A − B) =", P_A_menos_B)

    P_diferencia_simetrica = P["A"] + P["B"] - 2 * P["A∩B"]
    print("P(A Δ B) =", P_diferencia_simetrica)


propiedades_probabilidad()


```

```python
#3
import numpy as np

def problema_1():
    P_F = 0.40  # Probabilidad de que un paciente fume
    P_H_given_F = 0.75  # Probabilidad de que sea hombre dado que fuma
    P_M_given_no_F = 0.60  # Probabilidad de que sea mujer dado que no fuma
    P_H_given_no_F = 1 - P_M_given_no_F  # Probabilidad de que sea hombre dado que no fuma

    P_M = (1 - P_F) * 0.60 + P_F * (1 - P_H_given_F)
    print(f"P(M) = {P_M:.2f} (Mujer)")

    P_F_and_H = P_F * P_H_given_F
    print(f"P(F ∩ H) = {P_F_and_H:.2f} (Hombre y Fuma)")

    P_F_given_M = (P_F * (1 - P_H_given_F)) / P_M
    print(f"P(F | M) = {P_F_given_M:.2f} (Fume dado Mujer)")


def problema_2():
    lambdas = [1, 2, 3, 4]
    prior = [0.4, 0.3, 0.2, 0.1]
    x = 4
    lambda_hat = sum([l * p for l, p in zip(lambdas, prior)])
    print(f"a) Estimación inicial de λ = {lambda_hat:.2f}")

    def poisson_likelihood(x, lambda_i):
        return (lambda_i ** x) * np.exp(-lambda_i) / np.math.factorial(x)

    likelihoods = [poisson_likelihood(x, lambda_i) for lambda_i in lambdas]
    print(f"b) Verosimilitudes P(x|λ): {likelihoods}")

    posterior = [likelihood * prior_i for likelihood, prior_i in zip(likelihoods, prior)]
    normalizer = sum(posterior)
    posterior = [p / normalizer for p in posterior]
    print(f"c) Distribución posterior P(λ|x): {posterior}")

    print(f"d) Verificación de normalización: {sum(posterior):.2f}")

    max_posterior = max(posterior)
    best_lambda_index = posterior.index(max_posterior)
    print(f"e) El modelo que más probablemente explica los datos es λ = {lambdas[best_lambda_index]}")

    lambda_hat_posterior = sum([l * p for l, p in zip(lambdas, posterior)])
    print(f"f) Mejor estimación de λ = {lambda_hat_posterior:.2f}")


def problema_3():
    lambdas = [1, 2, 3, 4]
    prior = [0.4, 0.3, 0.2, 0.1]
    x1 = 4  # Primer intervalo
    x2 = 5  # Segundo intervalo

    lambda_hat = sum([l * p for l, p in zip(lambdas, prior)])
    print(f"a) Estimación inicial de λ = {lambda_hat:.2f}")

    def poisson_likelihood(x, lambda_i):
        return (lambda_i ** x) * np.exp(-lambda_i) / np.math.factorial(x)

    likelihoods_x1 = [poisson_likelihood(x1, lambda_i) for lambda_i in lambdas]
    likelihoods_x2 = [poisson_likelihood(x2, lambda_i) for lambda_i in lambdas]
    print(f"b) Verosimilitudes P(x|λ) para x1 y x2: {likelihoods_x1}, {likelihoods_x2}")

    posterior_x1 = [likelihood * prior_i for likelihood, prior_i in zip(likelihoods_x1, prior)]
    posterior_x2 = [likelihood * prior_i for likelihood, prior_i in zip(likelihoods_x2, prior)]

    normalizer_x1 = sum(posterior_x1)
    normalizer_x2 = sum(posterior_x2)

    posterior_x1 = [p / normalizer_x1 for p in posterior_x1]
    posterior_x2 = [p / normalizer_x2 for p in posterior_x2]

    posterior_combined = [p1 * p2 for p1, p2 in zip(posterior_x1, posterior_x2)]
    normalizer_combined = sum(posterior_combined)
    posterior_combined = [p / normalizer_combined for p in posterior_combined]

    print(f"c) Distribución posterior combinada P(λ|x1, x2): {posterior_combined}")

    lambda_hat_posterior = sum([l * p for l, p in zip(lambdas, posterior_combined)])
    print(f"d) Mejor estimación de λ = {lambda_hat_posterior:.2f}")


def problema_4():
    flujo_electron = 2
    flujo_proton = 1
    flujo_alpha = 1
    eficiencia_electron = 0.9
    eficiencia_proton = 0.6
    eficiencia_alpha = 0.2

    P_electron = flujo_electron / (flujo_electron + flujo_proton + flujo_alpha)
    P_proton = flujo_proton / (flujo_electron + flujo_proton + flujo_alpha)
    P_alpha = flujo_alpha / (flujo_electron + flujo_proton + flujo_alpha)

    prior = {'electron': P_electron, 'proton': P_proton, 'alpha': P_alpha}
    print(f"Prior Π(D|p): {prior}")


print("Problema 1:")
problema_1()
print("\nProblema 2:")
problema_2()
print("\nProblema 3:")
problema_3()
print("\nProblema 4:")
problema_4()
```

```python
#4
from math import factorial

def permutaciones_sin_repeticion(n, r):
    return factorial(n) // factorial(n - r)

def permutaciones_con_repeticion(n, r):
    return n ** r

def combinaciones_sin_repeticion(n, r):
    return factorial(n) // (factorial(r) * factorial(n - r))

def combinaciones_con_repeticion(n, r):
    return factorial(n + r - 1) // (factorial(r) * factorial(n - 1))

print("1. Podio de 3 personas (perm. sin repetición):", permutaciones_sin_repeticion(3, 2))

print("2. Ensalada de frutas (comb. sin repetición):", combinaciones_sin_repeticion(3, 2))

print("3. Cola de 5 amigos (perm. sin repetición):", factorial(5))

print("4. Premios para 8 participantes (perm. sin repetición):", permutaciones_sin_repeticion(8, 3))

print("5. Selección de 2 marineros de 10 (comb. sin repetición):", combinaciones_sin_repeticion(10, 2))

print("6. Acomodar 5 libros de 7 (perm. sin repetición):", permutaciones_sin_repeticion(7, 5))

print("7. Comité de 2 alumnos de 10 (comb. sin repetición):", combinaciones_sin_repeticion(10, 2))

print("8. Palabras de REMEMBER (perm. con repetición):", factorial(8) // (factorial(3) * factorial(2)))

print("9. Equipos con María (comb. sin repetición):", combinaciones_sin_repeticion(11, 5))

print("10. Jugos surtidos de 4 frutas (comb. sin repetición):",
      sum(combinaciones_sin_repeticion(4, r) for r in range(2, 5)))

print("11. Selección de presidente, vicepresidente y secretario (perm. sin repetición):",
      permutaciones_sin_repeticion(10, 3))

print("12. Premios de campeón y subcampeón (perm. sin repetición):", permutaciones_sin_repeticion(8, 2))

print("13. Números de 3 cifras distintas con dígitos del 1 al 7 (perm. sin repetición):",
      permutaciones_sin_repeticion(7, 3))

print("14. Números de 3 cifras con dígitos del 1 al 7 (perm. con repetición):", permutaciones_con_repeticion(7, 3))

print("15. Selección de comité de 3 estudiantes de 10 (comb. sin repetición):",
      combinaciones_sin_repeticion(10, 3))

print("16. Placas con 3 letras y 3 dígitos (perm. con repetición):", permutaciones_con_repeticion(26, 3) * permutaciones_con_repeticion(10, 3))

n = 7  # Cambiar según sea necesario
print("17. Maneras de sentar n personas alrededor de una mesa (perm. circular):", factorial(n - 1))

print("18. Combinaciones de helados con 3 sabores (comb. sin repetición):", combinaciones_sin_repeticion(7, 3))

print("19. Selección de gaseosas (comb. sin repetición):", combinaciones_sin_repeticion(6, 3),
      ", con repetición:", combinaciones_con_repeticion(6, 3))

# 20. Demostración de la fórmula de combinaciones con repetición
def demostrar_combinaciones_con_repeticion(n, r):
    formula = factorial(n + r - 1) // (factorial(r) * factorial(n - 1))
    print(f"Demostración: C({n},{r}) = {formula}")
    return formula

demostrar_combinaciones_con_repeticion(6, 3)
```
